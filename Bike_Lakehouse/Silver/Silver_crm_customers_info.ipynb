# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.types import StringType
from pyspark.sql.functions import trim,col
```

```python
RENAME_MAP = {
    "cst_id": "customer_id",
    "cst_key": "customer_key",
    "cst_firstname": "firstname",
    "cst_lastname": "lastname",
    "cst_marital_status": "marital_status",
    "cst_gndr": "gender",
    "cst_create_date": "created_date"
}
```

#Reading From Bronze

```python
df= spark.table("workspace_1.bronze.crm_cust_info")
```

# Data Transformation

# Trimming

```python
for field in df.schema.fields: # For Loops
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name, trim(col(field.name))) 
```

# Normalization

```python
df = (
    df
    .withColumn(
        "cst_marital_status",
        F.when(F.upper(F.col("cst_marital_status")) == "S", "Single")
        .when(F.upper(F.col("cst_marital_status")) == "M", "Married")
        .otherwise("n/a")
    )
    .withColumn(
        "cst_gndr",
        F.when(F.upper(F.col("cst_gndr")) == "M", "Male")
        .when(F.upper(F.col("cst_gndr")) == "F", "Female")
        .otherwise("n/a")
    )
)
```

# Renaming Columns

```python
for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

# Write Into Silver Table

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("workspace_1.silver.crm_customers")
)
```
